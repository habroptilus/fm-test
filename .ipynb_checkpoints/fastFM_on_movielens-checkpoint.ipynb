```python
%matplotlib inline
import numpy as np
import pandas as pd
logs = pd.read_csv("./ml-100k/u.data", names=["uid", "mid", "rating", "timestamp"], sep="\t")
y=np.array(logs["rating"])
logs=logs.drop(["timestamp","rating"], axis=1)
logs.tail(10)
```

```python
users = pd.read_csv("./ml-100k/u.user", names=["uid", "age" , "gender", "occupation" ,"zip_code"], sep="|")
users=users.drop(["age","zip_code"], axis=1)
users.tail(10)
```

```python
# itemのcontext追加情報の読み込み
#read_csvだとUnicodeDecodeErrorが起きてしまったのでhttps://qiita.com/niwaringo/items/d2a30e04e08da8eaa643で解決
# 後ろのcolumnはジャンルを表す.(複数ジャンルに属するitemもあるらしい)

import codecs
with codecs.open("./ml-100k/u.item", "r", "Shift-JIS", "ignore") as f:
    items = pd.read_table(f,names=["mid" , "title" , "released", "video_released", "IMDb_URL" ,
                                               "unknown" , "Action" , "Adventure" , "Animation" ,
                                               "Children" , "Comedy" ,"Crime" , "Documentary" , "Drama" , "Fantasy" ,
                                               "Film_Noir" ,"Horror" , "Musical" , "Mystery", "Romance" , "Sci_Fi",
                                               "Thriller", "War", "Western"],delimiter="|")
items=items.drop(["title", "released","video_released", "IMDb_URL" ],axis=1)
items.tail(10)
```

```python
log_user=pd.merge(logs,users,on="uid")
log_user_dummied=pd.get_dummies(log_user)
merged_df= pd.merge(log_user_dummied,items,on="mid")
features_df=pd.get_dummies(merged_df,columns=["uid","mid"])
features_df.tail(10)
```

```python
from  scipy.sparse import csr_matrix
X=csr_matrix(features_df.values)
```

```python
X.toarray().shape
```

```python
from sklearn.model_selection import train_test_split
test_rate=0.1
valid_rate=0.1

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=test_rate)
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train,test_size=valid_rate)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

def get_rmse(model,X, y):
    y_pred=model.predict(X)
    return np.sqrt(mean_squared_error(y_pred, y))

def train(model,X_train,y_train,X_valid,y_valid,n_iter,iter_size=1):
    model.fit(X_train,y_train) # initのためにこれが必要
    rmse_hist={"train":[],"valid":[]}
    iter_size=1
    for i in range(n_iter):
        model.fit(X_train, y_train,n_more_iter=iter_size)
        rmse_train=get_rmse(model,X_train,y_train)
        rmse_valid=get_rmse(model,X_valid,y_valid)
        rmse_hist["train"].append(rmse_train)
        rmse_hist["valid"].append(rmse_valid)
        if (i+1)%10==0: 
            print(f"Train : {rmse_train: 0.4f}, Valid : {rmse_valid:0.4f}")
    return model,rmse_hist
```

```python
from fastFM import als,sgd
model = als.FMRegression(n_iter=0, l2_reg_w=0.1, l2_reg_V=0.1, rank=4)

model,history=train(model,X_train,y_train,X_valid,y_valid, n_iter=100)
```

```python
y_pred=model.predict(X_test)
```

```python
y_train_mean=y_train.mean()*np.ones(len(y_test))
```

```python
print(f"[FM] : RMSE = {get_rmse(model,X_test,y_test):0.4f}")
print(f"[Baseline] : RMSE = {np.sqrt(mean_squared_error(y_train_mean, y_test)):0.4f}")
```

```python
import matplotlib.pyplot as plt

def plot_learning_curve(history):
    plt.figure()
    plt.title("Learning Curve")
    plt.xlabel("n_iter")
    plt.ylabel("RMSE")
    n_iter=len(history["train"])
    # Traing score と Test score をプロット
    plt.plot(list(range(1,n_iter+1)), history["train"], label="train")
    plt.plot(list(range(1,n_iter+1)), history["valid"], label="valid")
    plt.legend()
    plt.show()
```

```python
plot_learning_curve(history)
```
